# Drop Unnamed column

```python
dataset= dataset.drop("Unnamed: 0", axis=1)
print(dataset.head())
```

# Gender distribution

```python
plt.figure(figsize= (5,5))
a = sns.countplot(data = dataset, x="Gender")
a.bar_label(a.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#### From the above chart we get the insight about gender column that the number of females in the data is more that the males

```python
gb = dataset.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
print(gb)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's education and Student's score")
plt.show()
```

#### We check the relationship between Parents education and students scores and got the result that education of the parents have a good impact on students score

```python
gb1 = dataset.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean',"WritingScore": 'mean'})
print(gb1)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb1, annot = True)
plt.title("Relation between parent's marital status and student's score")
plt.show()
```

#### from the above heatmap we have got the conclusion that there is no impact on the student's score due to parent's marital status

# Ditribution of ethnic groups

```python
groupA = dataset.loc[(dataset["EthnicGroup"] == "group A")].count()
groupB = dataset.loc[(dataset["EthnicGroup"] == "group B")].count()
groupC = dataset.loc[(dataset["EthnicGroup"] == "group C")].count()
groupD = dataset.loc[(dataset["EthnicGroup"] == "group D")].count()
groupE = dataset.loc[(dataset["EthnicGroup"] == "group E")].count()

mylist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

l = ["group A", "group B", "group C", "group D", "group E"]

plt.pie(mylist, labels=l, autopct="%1.2f%%")
plt.title("Distribution of Ethnic groups")
plt.show()
```
